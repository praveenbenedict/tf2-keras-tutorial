```python
import os 
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '2'

import tensorflow as tf
physical_devices = tf.config.list_physical_devices('GPU')
print('Physical Devices', physical_devices)
tf.config.experimental.set_memory_growth(physical_devices[0], True)
```

### Initialization of tensors

```python
x = tf.constant(4)
print(x)
```

```python
x = tf.constant(4., shape=(1, 1))
print(x)
```

```python
x = tf.constant(4., shape=(1, 1), dtype=tf.float32)
print(x)
```

```python
x = tf.constant([
    [1, 2, 4],
    [4, 5, 6]
])
print(x)
```

```python
x = tf.constant([
    [1, 2, 4],
    [4, 5, 6]
], shape=(1, 2, 3))
print(x)
```

```python
x = tf.ones((3, 3))
print(x)
```

```python
x = tf.zeros((2, 3))
print(x)
```

```python
x = tf.eye(3)
print(x)
```

```python
x = tf.random.normal((2, 3))
print(x)
```

```python
x = tf.random.normal((2, 3), mean=0, stddev=1)
print(x)
```

```python
x = tf.random.uniform((1, 3), minval=0, maxval=1)
print(x)
```

```python
x = tf.range(9)
print(x)
```

```python
x = tf.range(start=1, limit=10, delta=2)
print(x)
```

```python
x = tf.cast(x, dtype=tf.float64)
print(x)
```

tf.float -> 16, 32, 64 ... tf.int -> 8, 16, 42, 64 ... tf.bool


### Mathematical operations

```python
x = tf.constant([1, 2, 3])
y = tf.constant([9, 8, 7])
```

```python
z = tf.add(x, y)
print(z)
```

```python
z = x + y
print(z)
```

```python
z = tf.subtract(x, y)
print(z)
```

```python
z = x - y
print(z)
```

```python
z = tf.divide(x, y)
print(z)
```

```python
z = x/y
print(z)
```

```python
z = tf.multiply(x, y)
print(z)
```

```python
z = x * y
print(z)
```

```python
z = tf.tensordot(x, y, axes = 1)
print(z)
```

```python
z = tf.reduce_sum(x*y, axis=0)
print(z)
```

```python
z = x**5
print(z)
```

```python
x = tf.random.normal((2, 3))
y = tf.random.normal((3, 4))
```

```python
z = tf.matmul(x, y)
print(z)
```

```python
z = x @ y
print(z)
```

### Indexing

```python
x = tf.constant([0, 1, 1, 2, 3, 1, 2, 3])
```

```python
print(x[:])
```

```python
print(x[1:])
```

```python
print(x[1:3])
```

```python
print(x[::2])
```

```python
print(x[::-1])
```

```python
indices = tf.constant([0, 3])
x_ind = tf.gather(x, indices)
print(x_ind)
```

```python
x = tf.constant([
    [1, 2],
    [3, 4], 
    [5, 6]
])
```

```python
print(x[0, :])
```

```python
print(x[0:2, :])
```

```python
print(x[0:2])
```

### Reshaping

```python
x = tf.range(9)
```

```python
print(x)
```

```python
x_ = tf.reshape(x, (3, 3))
print(x_)
```

```python
x_t = tf.transpose(x, perm=[1, 0])
print(x_t)
```
